```python
import pandas as pd
from pytrends.request import TrendReq

# Initialize a TrendReq object
pytrends = TrendReq(hl='en-US', tz=360)

import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
pytrends.build_payload(kw_list=["semiconductors"])
data = pytrends.interest_by_region()

# Sort the data by the keyword interest in descending order and select the top 10
top_regions = data.sort_values(by="semiconductors", ascending=False).head(20)

print("-----Top regions for 'semiconductors------'")
print(top_regions)
```

```python
# Prepare plot data
regions = top_regions.index.tolist()  # X-axis Labels (regions)
interest_levels = top_regions['semiconductors'].tolist()  # Heights of the bars

# Colors and labels 
colors = ['grey'] * 4  # Extend colors if more than 5 regions
labels = regions  # Use region names as labels

#font
font = {'family':'sans-serif',
        'weight': 'regular',
        'size' : 50}
rc('font', **font)  # Apply font settings globally

# Create bar chart
plt.figure(figsize=(80, 40))
plt.bar(regions, interest_levels, color=colors, tick_label=labels, width=0.5)


# Adding titles and labels
plt.title('Interest in "Semiconductors" by Region', fontsize=80)
plt.xlabel('Region', fontsize=50)
plt.ylabel('Interest Level', fontsize=50)
plt.xticks(rotation=45, fontsize=40, ha= 'right')  # Rotate x-axis labels for better readability
#plt.yticks(fontsize=24)

# Show the plot
plt.show()
```

```python
#Related Queries
related_queries = pytrends.related_queries()
top_related_queries = related_queries['semiconductors']['top']


print("\n-----Top related queries for 'semiconductors'-----")
print(top_related_queries)
```

```python
font = {'family':'sans-serif',
        'weight': 'regular',
        'size' : 50}
rc('font', **font)

if top_related_queries is not None and not top_related_queries.empty:
    # Prepare plot data
    queries = top_related_queries['query']
    values = top_related_queries['value']

    plt.figure(figsize=(80, 40))
    plt.bar(queries, values, color='grey', width=0.5)
    plt.xlabel('queries', fontsize=60)
    plt.ylabel('Popularity', fontsize=60)
    plt.title('Top Related Queries for "Semiconductors"', fontsize=80)
    plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability
    plt.tight_layout()  # Adjust layout to make room for label rotation
    plt.show()
else:
    print("No data available to plot.")
```
